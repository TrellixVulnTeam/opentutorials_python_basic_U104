## 조건 연산자





- x in container : x가 콘테이너 안에 존재하는가? 존재 : True
- x not in container : x가 콘테이너 안에 존재하는가? 존재하지 않음 : False

** container => 리스트 튜플 문자열 등의 배열을 말함

```python
1 in [1,2,3]
```

```python
1 not in [1,2,3]
```

```python
'a' in ('a','b','c')
```

```python
'j' not in 'python'
```

### 만약 조건문에서 아무일도 하지 않게 하고 싶다면?

- 조건문 안에서 pass를 사용한다.

```python
pocket=['paper','money','cellphone']
```

```python
if 'money' in pocket:
    pass
else:
    print("카드를 꺼내라")
```

```python
# 포켓리스트 안에 money가 있기 때문에 if문 다음 문장인 pass가 수행되고 아무 결과 값도 보여주지 않는다.
```
