## while문 강제로 빠져나가기 : Break

```python
coffee = 10
money = 300
```

```python
while money :
    print("돈을 받았으니 커피를 줍니다")
    coffee=coffee-1
    print("남은 커피의 양은 %d개 입니다." %coffee)
    if coffee == 0 :
        print("커피가 다 떨어졌으니 판매를 중지합니다.")
        break
```

## while문의 처음으로 돌아가기 : Continue

while문 안의 문장을 수행할 때 입력 조건을 검사해서 조건에 맞지 않으면 while문을 빠져나간다.   
그런데 프로그래밍을 하다 보면 while문을 빠져나가지 않고 while문의 맨 처음(조건문)으로 다시 돌아가게 만들고 싶은 경우가 생기게 된다.    
이때 사용하는 것이 바로 continue문이다.  

```python
a=0
```

```python
while a<10:
    a=a+1
    if a%2 == 0: # 2의 배수일 경우 if문이 실행
        continue # continue 이하의 명령을 실행하지 않고 반복문의 처음으로 돌아간다.
    print(a)
```
