## 리스트 복사

```python
a=[1,2,3]
b=a
a
```

```python
b
```

```python
# b 변수에 a 변수를 대입하면 어떻게 될까? b와 a는 같은 걸까 다른 걸까? 결론부터 말하면 b는 a와 완전히 동일하다고 할 수 있다. 
# 다만 [1, 2, 3] 리스트를 참조하는 변수가 a 변수 1개에서 b 변수가 추가되어 2개로 늘어났다는 차이만 있을 뿐이다.

id(a)
```

```python
id(b)
# 리스트가 늘어나는 것이 아니라 해당 리스트를 참조하는 변수가 늘어나는 것이다.
```

```python
# a 와 b가 가리키는 객체는 동일하다
a is b # 동일한 객체를 가리키는 지 판단하는 명령어 is
```

```python
# 차이점 보기
a=[1,2,3]
b=[1,2,3]
```

```python
a is b # 변수 a와 b는 서로다른 메모리를 가리킴 복사만 같은 메모리에 참조변수가 늘어나는 것이다.
```

```python
b
```

- copy모듈 사용하여 복사하기

```python
from copy import copy
```

```python
a = [1,2,3]
```

```python
b=copy(a) # b=a[:] , b=a 와 동일
```

```python
b
```

## 변수를 만드는 여러가지 방법

```python
a
```

```python
b
```

```python
# 리스트로 변수 선언
[a, b]=['python', 'life']
```
